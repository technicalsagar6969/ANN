4. Realization of logical AND function using Perceptron learning rule.

```python
import numpy as np

class Perceptron:
    def __init__(self, num_inputs, learning_rate=0.1, epochs=100):
        self.num_inputs = num_inputs
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = np.random.rand(num_inputs)
        self.bias = np.random.rand()

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        weighted_sum = np.dot(inputs, self.weights) + self.bias
        return self.activation(weighted_sum)

    def train(self, training_inputs, labels):
        for _ in range(self.epochs):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                error = label - prediction
                self.weights += self.learning_rate * error * inputs
                self.bias += self.learning_rate * error

if __name__ == "__main__":
    # Define training data (inputs and corresponding labels)
    training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    labels = np.array([0, 0, 0, 1])

    # Create a Perceptron with 2 input neurons
    perceptron = Perceptron(num_inputs=2)

    # Train the Perceptron on the training data
    perceptron.train(training_inputs, labels)

    # Test the trained Perceptron
    test_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    print("Input\t\tPredicted Output")
    for inputs in test_inputs:
        prediction = perceptron.predict(inputs)
        print(f"{inputs}\t\t{prediction}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models

# Define the model architecture
def build_model(input_shape, num_classes):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# Load the dataset (replace this with your own dataset loading code)
def load_data():
    # Sample data loading
    x_train = np.random.rand(100, 28, 28, 1)  # Sample input data
    y_train = np.random.randint(0, 10, size=(100,))  # Sample labels
    x_test = np.random.rand(20, 28, 28, 1)  # Sample input data for testing
    y_test = np.random.randint(0, 10, size=(20,))  # Sample labels for testing
    return (x_train, y_train), (x_test, y_test)

# Main function to train, save, and evaluate the model
def main():
    # Load the dataset
    (x_train, y_train), (x_test, y_test) = load_data()

    # Preprocess the data (if needed)

    # Define input shape and number of classes
    input_shape = x_train[0].shape
    num_classes = len(np.unique(y_train))

    # Build the model
    model = build_model(input_shape, num_classes)
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    model.fit(x_train, y_train, epochs=10, batch_size=32)

    # Save the trained model
    model.save('speech_recognition_model.h5')
    print("Model saved successfully.")

    # Evaluate the model
    test_loss, test_acc = model.evaluate(x_test, y_test)
    print('Test accuracy:', test_acc)

# Run the main function
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the saved model
model = load_model('speech_recognition_model.h5')
print("Model loaded successfully.")

# Function to preprocess audio data
def preprocess_audio(audio_file):
    # Load the audio file
    audio, _ = librosa.load(audio_file, sr=16000)  # Sampling rate of 16kHz

    # Extract features (e.g., MFCCs)
    mfccs = librosa.feature.mfcc(y=audio, sr=16000, n_mfcc=40)

    # Resize the features to match the input shape of the model
    resized_mfccs = np.resize(mfccs, (28, 28))

    # Expand dimensions to create a channel dimension
    input_data = np.expand_dims(resized_mfccs, axis=-1)

    return input_data.reshape(1, 28, 28, 1)


# Function to perform speech recognition
def recognize_speech(audio_file):
    # Preprocess the audio data
    processed_audio = preprocess_audio(audio_file)

    # Make predictions using the model
    prediction = model.predict(processed_audio)

    # Decode the predicted label (replace this with your label decoding logic)
    # Here, we assume the labels are integers from 0 to 9 representing digits
    decoded_label = np.argmax(prediction)

    return decoded_label

# Path to the audio file you want to recognize
audio_file_path = 'harvard.wav'

# Perform speech recognition
predicted_label = recognize_speech(audio_file_path)
print("Predicted label:", predicted_label)
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Function to fetch historical stock data
def fetch_stock_data(symbol, start_date, end_date):
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data

# Function to preprocess stock data
def preprocess_stock_data(data):
    # Extracting 'Close' prices
    close_prices = data['Close'].values.reshape(-1, 1)

    # Scaling the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(close_prices)

    return scaled_data, scaler

# Function to prepare input features and target variable
def prepare_data(data, lookback):
    X, y = [], []
    for i in range(len(data) - lookback):
        X.append(data[i:(i + lookback), 0])
        y.append(data[i + lookback, 0])
    return np.array(X), np.array(y)

# Define the parameters
symbol = 'AAPL'  # Stock symbol (Apple Inc. in this example)
start_date = '2010-01-01'  # Start date for historical data
end_date = '2022-01-01'  # End date for historical data
lookback = 30  # Number of previous days' prices to consider for prediction

# Fetch historical stock data
stock_data = fetch_stock_data(symbol, start_date, end_date)

# Preprocess stock data
scaled_data, scaler = preprocess_stock_data(stock_data)

# Prepare input features and target variable
X, y = prepare_data(scaled_data, lookback)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Define the feedforward neural network model
model = Sequential([
    Dense(64, activation='relu', input_shape=(lookback,)),
    Dense(64, activation='relu'),
    Dense(1)  # Output layer (single neuron for regression)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# Evaluate the model
test_loss = model.evaluate(X_test, y_test)
print('Test Loss:', test_loss)

# Make predictions
predictions = model.predict(X_test)

# Inverse transform the predictions
predictions = scaler.inverse_transform(predictions)

# Print the first few predictions
print("Predictions:")
print(predictions[:5])
```
